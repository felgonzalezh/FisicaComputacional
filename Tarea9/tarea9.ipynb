```python
%pylab inline
#load the list of points
x1_obs = 0
y1_obs = 0
z1_obs = 0
x2_obs = 0
y2_obs = 0
z2_obs = 0

obs_data = loadtxt("traces.dat")
x1_obs = obs_data[:,0]
y1_obs = obs_data[:,1]
z1_obs = obs_data[:,2]
x2_obs = obs_data[:,3]
y2_obs = obs_data[:,4]
z2_obs = obs_data[:,5]
```

```python
def likelihood(x2_obs, x2_model,y2_obs, y2_model):
    delta_x=x2_obs-x2_model
    delta_y=y2_obs-y2_model
    chi_squared = (-1.0)*sum((delta_x*delta_x)+(delta_y*delta_y))
    return chi_squared
```

```python
def my_model_x(x1_ob, x0, z0, z1_ob, z2_ob):
    m_xz=(x1_ob-x0)/(z1_ob-z0)
    x_pred=(m_xz*(z2_ob-z0))+x0
    return x_pred
def my_model_y(y1_ob, y0, z0, z1_ob, z2_ob):
    m_yz=(y1_ob-y0)/(z1_obs-z0)
    y_pred=(m_yz*(z2_ob-z0))+y0
    return y_pred
```

```python
x0_walk = empty((0)) #this is an empty list to keep all the steps
y0_walk = empty((0))
z0_walk = empty((0))



x0_walk = append(x0_walk, random.random())
y0_walk = append(y0_walk, random.random())
z0_walk = append(z0_walk, (2900*(random.random())-3000))
print x0_walk
print y0_walk
print z0_walk
```

```python
n_iterations = 5000 #this is the number of iterations I want to make
for i in range(n_iterations):
    x0_prime = np.random.normal(x0_walk[i], 1.0) 
    y0_prime = np.random.normal(y0_walk[i], 1.0)
    z0_prime = np.random.normal(z0_walk[i], 200.0) 
    
    while z0_prime > 0.0 or z0_prime < -5000.0:
        z0_prime = np.random.normal(z0_walk[i], 300)
  
    x_init = my_model_x(x1_obs, x0_walk[i], z0_walk[i], z1_obs, z2_obs)
    y_init = my_model_y(y1_obs, y0_walk[i], z0_walk[i], z1_obs, z2_obs)
    
    x_prime = my_model_x(x1_obs, x0_prime, z0_prime, z1_obs, z2_obs)
    y_prime = my_model_y(y1_obs, y0_prime, z0_prime, z1_obs, z2_obs)
    #print x_prime
    
    alpha = likelihood(x2_obs, x_prime, y2_obs, y_prime)-likelihood(x2_obs, x_init, y2_obs, y_init)
    #print likelihood(x2_obs, x_prime, y2_obs, y_prime)
    #print likelihood(x2_obs, x_init, y2_obs, y_init)
    #print alpha
    if(alpha>0.0):
    
        x0_walk  = append(x0_walk,x0_prime)
        y0_walk  = append(y0_walk,y0_prime)
        z0_walk  = append(z0_walk,z0_prime)
    else:
        beta = random.random()
        #print beta
        #print exp(alpha)
        if(beta<exp(alpha)):
            x0_walk = append(x0_walk,x0_prime)
            y0_walk = append(y0_walk,y0_prime)
            z0_walk = append(z0_walk,z0_prime)
        else:
            x0_walk = append(x0_walk,x0_walk[i])
            y0_walk = append(y0_walk,y0_walk[i])
            z0_walk = append(z0_walk,z0_walk[i])
```
